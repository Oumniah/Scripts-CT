```python
#consolidations 2021 , it works!
import pandas as pd
from datetime import date, datetime, timedelta

errors = []

#prepare date

def datespan(startDate, endDate, delta=timedelta(days=1)):
    currentDate = startDate
    while currentDate < endDate:
        yield currentDate
        currentDate += delta

weeks = []    
day_start = []
day_end = []

for day in datespan(date(2020, 12, 30), date(2021, 1, 7), delta=timedelta(days=7)):

    day_start.append(day.strftime("%d-%m-%Y"))
    #print(day.strftime("%d-%m-%Y"))

for day in datespan(date(2021, 1, 5), date(2021, 1, 13), delta=timedelta(days=7)):

    day_end.append(day.strftime("%d-%m-%Y"))
    #print(day.strftime("%d-%m-%Y"))

#print(day_end)
for i in range(0,len(day_end)):
    weeks.append("du "+day_start[i]+" au "+day_end[i])

for i in weeks :
    try:
        file = "C:/Users/ohoummi/Desktop/Oumnia/Rapport H recette 2021/Consolidation recettes Transport "+i+".xlsx"
        data = pd.read_excel(file, header=11) 
        if list(data.columns)[0]=="Unnamed: 0" :
            data = data.drop(columns = ["Unnamed: 0"])
        #print(data)
        data.columns = ['date','v_dtt_support','v_dtt_charge','v_dtt_ab_hebdo','v_dtt_ab_mens',\
                                            'v_dtt_ab_mens_etud','v_agence_support_b','v_agence_support_c','v_agence_voyage',\
                                            'v_agence_ab_hebdo','v_agence_ab_mens',\
                                            'v_agence_ab_mens_etud','v_depo_support_b','v_depo_support_c','v_depo_voyage',\
                                            'v_depo_ab_hebdo','v_depo_ab_mens','v_depo_ab_mens_etud',\
                                            'v_pv_support_b','v_pv_voyage',\
                                            'ca_dtt_support','ca_dtt_charge','ca_dtt_ab_hebdo','ca_dtt_ab_mens',\
                                            'ca_dtt_ab_mens_etud','ca_dtt_banque',\
                                            'ca_agence_support_b','ca_agence_support_c','ca_agence_voyage',\
                                            'ca_agence_ab_hebdo','ca_agence_ab_mens',\
                                            'ca_agence_ab_mens_etud','ca_agence_banque',\
                                            'ca_depo_support_b','ca_depo_support_c','ca_depo_voyage',\
                                            'ca_depo_ab_hebdo','ca_depo_ab_mens','ca_depo_ab_mens_etud',\
                                            'ca_pv_support_b','ca_pv_voyage','rembours_banque','total']


        #get the file
        #data = pd.read_excel(file, header=11,names=columns, dtype={'date' : str}) 


        #lists initialization
        e_dtt_support = []
        e_dtt_charge = []
        e_dtt_ab_hebdo = []
        e_dtt_ab_mens = []
        e_dtt_ab_mens_etud = []
        e_agence_support_b = []
        e_agence_support_c = []
        e_agence_voyage = []
        e_agence_ab_hebdo = []
        e_agence_ab_mens = []
        e_agence_ab_mens_etud = []
        e_depo_support_b = []
        e_depo_support_c = []
        e_depo_voyage = []
        e_depo_ab_hebdo = []
        e_depo_ab_mens = []
        e_depo_ab_mens_etud = []
        e_pv_support_b = []
        e_pv_voyage = []

        #gabs calculation
        e_dtt_support = data['ca_dtt_support'] - (data['v_dtt_support']*2)
        e_dtt_charge = data['ca_dtt_charge'] - (data['v_dtt_charge']*6)
        e_dtt_ab_hebdo = data['ca_dtt_ab_hebdo'] - (data['v_dtt_ab_hebdo']*60)
        e_dtt_ab_mens = data['ca_dtt_ab_mens'] - (data['v_dtt_ab_mens']*230)
        e_dtt_ab_mens_etud = data['ca_dtt_ab_mens_etud'] - (data['v_dtt_ab_mens_etud']*150)
        e_agence_support_b = data['ca_agence_support_b'] - (data['v_agence_support_b']*2)
        e_agence_support_c = data['ca_agence_support_c'] - (data['v_agence_support_c']*15)
        e_agence_voyage = data['ca_agence_voyage'] - (data['v_agence_voyage']*6)
        e_agence_ab_hebdo = data['ca_agence_ab_hebdo'] - (data['v_agence_ab_hebdo']*60)
        e_agence_ab_mens = data['ca_agence_ab_mens'] - (data['v_agence_ab_mens']*230)
        e_agence_ab_mens_etud = data['ca_agence_ab_mens_etud'] - (data['v_agence_ab_mens_etud']*150)
        e_depo_support_b = data['ca_depo_support_b'] - (data['v_depo_support_b']*2)
        e_depo_support_c = data['ca_depo_support_c'] - (data['v_depo_support_c']*15)
        e_depo_voyage = data['ca_depo_voyage'] - (data['v_depo_voyage']*6)
        e_depo_ab_hebdo = data['ca_depo_ab_hebdo'] - (data['v_depo_ab_hebdo']*60)
        e_depo_ab_mens = data['ca_depo_ab_mens'] - (data['v_depo_ab_mens']*230)
        e_depo_ab_mens_etud = data['ca_depo_ab_mens_etud'] - (data['v_depo_ab_mens_etud']*150)
        e_pv_support_b = data['ca_pv_support_b'] - (data['v_pv_support_b']*2)
        e_pv_voyage = data['ca_pv_voyage'] - (data['v_pv_voyage']*6)

        print(data)
        #dataframe making

        d = {
            'Date' : data['date'],
            'Dtt Ecart support' : e_dtt_support,
            'Dtt chargement' : e_dtt_charge,
            'Dtt Ecart Ab Hebdo' : e_dtt_ab_hebdo,
            'Dtt Ecart Ab Mens' : e_dtt_ab_mens,
            'Dtt Ecart Ab Mens Etud' : e_dtt_ab_mens_etud,
            'Agence Ecart support BSC' : e_agence_support_b,
            'Agence Ecart support CSC' : e_agence_support_c,
            'Agence Ecart voyage' : e_agence_voyage,
            'Agence Ecart Ab Hebdo' : e_agence_ab_hebdo,
            'Agence Ecart Ab Mens' : e_agence_ab_mens,
            'Agence Ecart Ab Mens Etud' : e_agence_ab_mens_etud,
            'Depo Ecart support BSC' : e_depo_support_b,
            'Depo Ecart support CSC' : e_depo_support_c,
            'Depo Ecart voyage' : e_depo_voyage,
            'Depo Ecart Ab Hebdo' : e_depo_ab_hebdo,
            'Depo Ecart Mens' : e_depo_ab_mens,
            'Depo Ecart Mens Etud' : e_depo_ab_mens_etud,
            'PV Ecart support BSC' : e_pv_support_b,
            'PV Ecart support voyage' : e_pv_voyage
        }

        #file generation
        result = pd.DataFrame(data = d)
        result.to_excel('Ecart '+i+'.xlsx',index=False)
        print("file generated !")

    except Exception as e :
        errors.append(e)
        print(i)
        continue 
```
